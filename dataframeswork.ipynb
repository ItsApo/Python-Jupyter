Aqui leo los archivos y los importo a cada dataframe

```python
DFjson = pd.read_json('import.json')
```

```python
DFcsv = pd.read_csv('importar.csv')
```

```python
DFxlsx = pd.read_excel('importeu.xlsx')
```

Este CombinedDataFrame sirve para combinar los 3 dataframes que cree y que se genere uno solo con todos los datos de todos los archivos

```python
CombinedDataFrame = pd.concat([DFjson, DFcsv, DFxlsx])
```

Aqui imprimo el dataframe que contiene los datos de los 3 archivos (o 3 dataframes)

```python
print(CombinedDataFrame)
```

Y aqui por ultimo exporto los datos en un archivo JSON para que queden guardados en los archivos listo para descargarse

```python
CombinedDataFrame.to_json('ExportData.json', orient='records', lines=True)
```
